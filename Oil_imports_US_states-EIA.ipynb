```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
all_grades = pd.read_csv("imports_clean.csv", header=0, index_col=0)
#all_grades = all_grades.drop(['Pinned Series'])
```

```python
all_grades = all_grades.drop('World', axis=0)
all_grades.replace('--', np.nan).dropna(how='all');
all_grades[:5]
```

```python
all_grades = all_grades.sort_values(by=['X2016'])
all_grades = all_grades.T
all_grades[:5]
```

```python
all_grades.plot.bar(stacked=True, legend=False);
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```
